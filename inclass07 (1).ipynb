```python
import numpy as np
import matplotlib.pyplot as plt

# 1차원 입력 ( + bias를 포함)
x = 2 * np.random.rand(100, 1)

# 1차원 출력 = theta_0 + theta_1 * x
y = 4 + 5*x + np.random.randn(100,1) #np.random~은 노이즈를 섞은 것. 이걸 안 쓰면 올곧은 모양으로 나옴

plt.figure(figsize = (5,3))
plt.scatter(x,y,s=10)
plt.xlabel('x')
plt.ylabel('y')
#그림에서 가로축은 입력, 세로축은 출력이 나옴
```

```python
# analytic, normal equation

# design matrix = 샘플 수 x 입력 수 (차원의 수)
X = np.hstack([np.ones((100,1)), x]) # 100 x 2 형태 #첫번째 열에는 1로 가득찬 열벡터 생성
# X = x #bias를 포함하고 싶지 않을 때 윗줄을 생략하고 얘를 씀. 입력을 바로 X에 넣음

# \hat{theta} = (X^T X)^{-1} X^T y
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

x_new = np.linspace(0,2,10).reshape(-1,1)
# y_new = theta[0,0]*x_new
y_new = theta[0,0] + theta[1,0]*x_new

plt.figure(figsize = (5,3))
plt.scatter(x,y,s=10)
plt.plot(x_new,y_new)
plt.xlabel('x')
plt.ylabel('y')
```
