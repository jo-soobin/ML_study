```python
import numpy as np
import matplotlib.pyplot as plt
from pytictoc import TicToc
import math
```

```python
x = np.linspace(0, 1, 10)
y = 2*x + 1

plt.figure(figsize = (5, 5)) # 빈 창을 하나 생성, figsize = (가로크기, 세로크기)
plt.subplot(2, 1, 1) # subplot(나눌 행의 수, 나눌 열의 수, 몇번째 위치)
plt.plot(x, y) # 선으로, 순서대로 이어서, plot(가로축, 세로축)

plt.subplot(2, 1, 2)
plt.scatter(x, y) # 산포도, 점으로, scatter(가로축, 세로축)
```

```python
theta = np.linspace(0, 2*math.pi, 100)
x = np.cos(theta) # 라디안으로 넣어야함
y = np.sin(theta)

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(theta, x) # 앞에 있는건 아래에 깔리는 그림
plt.plot(theta, y) # 자동으로 겹쳐서, 뒤에 있는걸 위에 그림

plt.subplot(2, 1, 2)
plt.plot(x, y)
```

```python
# 랜덤으로 샘플을 생성하는 것
x = np.random.rand(10000, 2) # uniform distribution [0, 1], np.random.rand(행, 열)
y = np.random.randn(10000, 2) # normal distribution (표준정규분포 N(0,1)), np.random.randn(행, 열)

plt.figure(figsize = (5, 10))
plt.subplot(2, 1, 1)
plt.scatter(x[:,0], x[:,1])

plt.subplot(2, 1, 2)
plt.scatter(y[:,0], y[:,1])
```

```python
# histogram
num_bins = 25 # 나눌 구간의 개수

plt.figure()
plt.subplot(2, 1, 1)
plt.hist(x[:,0], num_bins)

plt.subplot(2, 1, 2)
plt.hist(y[:,0], num_bins)
```

```python
# np.hstack # ndarray를, 가로방향으로 쌓아주세요, 행방향
# np.vstack # ndarray를, 세로방향으로 쌓아주세요, 열방향

x = np.linspace(0, 10, 6)
y = np.linspace(10, 20, 6)

print(x.shape)
print(y.shape)

z1 = np.hstack([x,y]) # numpy에서 벡터는, 행벡터가 기본
z2 = np.vstack([x,y])

print(z1.shape)
print(z2.shape)
print(z1)
print(z2)

X = np.zeros((3,4))
Y = np.ones((2,4))

Z = np.vstack([X,Y])
print(X)
print(Y)
print(Z)
```

```python
# 2차원, 100개의 랜덤 샘플을 생성을 함 = 샘플 X
# 2차원의 고정된 점 하나를 정해서 = 기준점 z
# 기준점으로부터, 샘플까지의 거리를 계산
# 계산한 거리를 기준으로, 일정 거리 이상이 되는 애들을 따로 표시

# np.where
# for문

t = TicToc()

N = 1000
D = 2
X = np.random.rand(N, D)
z = np.array([0.3, 0.4])

# 거리 구하기1 = for문으로
t.tic()
dist1 = np.zeros(N) # 벡터로 생성
for n in range(0, N):
    dist1[n] = np.sqrt((X[n,0] - z[0])**2 + (X[n,1] - z[1])**2)    
t.toc()

# 거리 구하기2 = for문 안쓰고, repmat + sum으로
t.tic()
Z = np.matlib.repmat(z.reshape(1,2), N, 1) # X랑 동일한 shape으로 생성
dist2 = np.sqrt(np.sum((X - Z)**2, axis = 1))
t.toc()
    
d_star = 0.2 # 거리의 기준
id_long = np.where(dist1 >= d_star)     
    
plt.figure(figsize = (5,5))
plt.scatter(X[id_long,0], X[id_long,1], s = 30, facecolor = 'cyan')
plt.scatter(X[:,0], X[:,1], s = 10, facecolor = 'gray')
plt.scatter(z[0], z[1], marker = 'x', facecolor = 'red')
```
