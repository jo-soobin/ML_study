```python
import numpy as np
import numpy.matlib
import math
import matplotlib.pyplot as plt

# 벡터의 차원
D = 2

# 랜덤으로 설정한 기준점
z = np.random.randn(D)

# 생성할 샘플의 개수
N = 100

# 기준이 되는 거리
d_star = 2

X = 10*(np.random.rand(N, D) - 0.5)
Z = np.matlib.repmat(z.reshape((1,D)), N, 1)
dist = np.sqrt(np.sum((X - Z)**2, axis = 1))
```

```python
dist_max = np.max(dist)
id_max = np.argmax(dist)

id_long = np.where(dist >= d_star)

theta = np.linspace(0,2*math.pi,100)
x1 = d_star*np.cos(theta) + z[0]
x2 = d_star*np.sin(theta) + z[1]

plt.figure(figsize = (5, 5))
ax = plt.subplot(1,1,1)
plt.plot(x1, x2, color = 'k')
plt.scatter(X[id_max,0], X[id_max,1], s = 100, marker = 'o', facecolor = 'w', edgecolor = 'b')
plt.scatter(X[id_long,0], X[id_long,1], s = 50, marker = 'o', facecolor = 'w', edgecolor = 'cyan')
plt.scatter(X[:,0], X[:,1], s = 10, facecolor = 'gray')
plt.scatter(z[0], z[1], marker = 'x', facecolor = 'r')
ax.set_aspect('equal')
plt.savefig('code1.eps')
```
